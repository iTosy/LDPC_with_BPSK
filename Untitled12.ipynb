```python
from matplotlib import pyplot as plt
```

```python
import numpy as np
from pyldpc import make_ldpc, encode, decode, get_message

```

```python
n = 1440
d_v = 95
d_c = 720
snr_db = 11
nrofblocks = 50
H, G = make_ldpc(n, d_v, d_c, systematic=True, sparse=True)
k = G.shape[1]
rate = k / n
nrof_bits = k*nrofblocks
source_bits = np.random.randint(2, size=nrof_bits)
ldpc_coded = np.random.randint(1, size=n*nrofblocks )
for i in range(nrofblocks):
  ldpc_coded[i*n:(i+1)*n] = np.matmul(source_bits[i*k:(i+1)*k],G.T)%2

 

x_radians = ldpc_coded*np.pi
tx_signal = np.cos(x_radians) + 1j*np.rint(np.sin(x_radians)) 
plt.plot(np.real(tx_signal), np.imag(tx_signal), '.')
plt.grid(True)
plt.show()


noise_variance = 1.0 / (rate * pow(10, 0.1 * snr_db))
noise = np.random.randn(len(ldpc_coded)) + 1j*np.random.randn(len(ldpc_coded))
noise *= np.sqrt(noise_variance)
rx_signal = tx_signal + noise

plt.plot(np.real(rx_signal), np.imag(rx_signal), '.')
plt.grid(True)
plt.show()


'''Demodulate the signal'''
demodulated_bits = np.random.randint(1, size=n*nrofblocks )
for i in range(len(demodulated_bits)):
  if (np.abs(rx_signal[i]-(1+0j)) < np.abs(rx_signal[i]-(-1+0j))):
    demodulated_bits[i] = 0
  else:
    demodulated_bits[i] = 1

decoded = np.random.randint(1, size=nrof_bits)

for i in range(nrofblocks):
  decoded[i*k:(i+1)*k] = get_message(G, demodulated_bits[i*n:(i+1)*n])
  errors=0
  berrors=0

for i in range(nrofblocks):
  f=0
  for j in range(k):
      if(decoded[i*k+j] != source_bits[i*k+j]):
        if(f==0):
          errors+=1
        berrors+=1
        f=1
        
ber = berrors/(nrofblocks*k)       
bler = errors/(nrofblocks)
print("bler for (%d,%d) ldpc is %f" %(n,k,bler))
print("ber for (%d,%d) ldpc is %f" %(n,k,ber))
```
